## Infographic Rusults:

#### What country had the most cybersecurity threat?

```python
df['Country'].value_counts().head(10).plot(kind = 'barh', figsize = (8, 5), color = 'red')
plt.title('Cybersecurity Threat per Country')
plt.ylabel('Country')
plt.xlabel('Amount of Threats')
plt.show()
```

#### What is the most commonly used attack type?

```python
df['Attack Type'].value_counts().head(10).plot(kind = 'barh', figsize = (6, 4), color = 'blue')
plt.title('Most Used Attack Type')
plt.xlabel('Amount of Attacks')
plt.ylabel('Attack Type')
plt.show()
```

#### Find the top 3 most targeted Industry.

```python

plt.figure(figsize=(15, 5))
plt.title('Top 3 Most Targeted Industries')
sns.histplot(df['Target Industry'])
plt.ylabel('Number of Incidents')
```

#### What Industry took the most Financial damage?

```python
plt.figure(figsize=(15, 5))
plt.title('Financial Damage per Industry')
plt.xlabel('Target Industry')
plt.ylabel('Financial Loss (in Million $)')
sns.barplot(df, x = 'Target Industry', y = 'Financial Loss (in Million $)', color='cyan', estimator= 'sum')
```

#### Were there any major discrepancies in the yearly cybersecurity attacks? 

```python
df['Year'].value_counts().sort_index().plot(kind = 'line', figsize = (13, 6), color = 'green')
plt.title('Cybersecurity Threat Timeline')
plt.ylabel('Amount of Threats')
plt.xlabel('Year')
plt.show()
```

#### Find any correlation between the types of attacks and each individual industry.

```python
industry_attack = pd.crosstab (df['Target Industry'], df['Attack Type'])
plt.figure(figsize =(12, 7))
sns.heatmap(industry_attack, cmap='Reds', linewidths=0.6, linecolor='Black')

plt.title('Correlation Metric Heatmap')
plt.xlabel('Attack Type')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

#### Average resolution time per defensive mechanism.

```python
top_def = df.groupby('Defense Mechanism Used')['Incident Resolution Time (in Hours)'].mean().sort_values()
display(top_def.head(10))
```

#### Top Incident Resolution in the IT industry.

```python
common = df.groupby(['Target Industry', 'Defense Mechanism Used']).agg({'Incident Resolution Time (in Hours)': 'mean', 'Defense Mechanism Used': 'count'}).rename(columns={'Defense Mechanism Used': 'Usage Count'}).reset_index()

top_def = common.sort_values(['Target Industry','Incident Resolution Time (in Hours)'])

defense = common[common['Target Industry'] == 'IT']
defense = defense.sort_values('Incident Resolution Time (in Hours)')
display(defense.head(10))
```

#### Are there any correlation between the type of industry and security vulnerability type?

```python
ind_def = pd.crosstab(df['Target Industry'], df['Security Vulnerability Type'])
display(ind_def.head(10))
```
